### Ciclos
El código se ejecuta n veces hasta que indiquemos un límite, ya sea una condición o un valor
Las bodades de un ciclo es que permite ejecutar un bloque de código aplicable a un conjunto de datos, es como cuándo tenemos una rutina diaria y ponemos en marcha nustro código mental, ya no se piensa que hacer, sino que se aplica la rutina.

#### Ciclo For

El ciclo For se utiliza habitualmente cuando sabemos la cantidad de veces que se requiere ejecutar un conjunto de intrucciones, por ejemplo se debe contar hasta 10.

Sintaxis

**for** variable **in** elemento iterable (lista, cadena, range, etc.)**:**
>`cuerpo del bucle, puede contener otros ciclos y condiciones if`


```python
# Ejmeplo_1: Ejecutar n veces un código, en este caso mostrando el contenido de la varia

veces = int(input("¿Cuántas quiere ejecutar el ciclo? "))
i=10 # Se asigna 10 para que observes que la variable i cuando se utiliza en for igual comienza en i

for i in range(veces):
    print("La varible iteradora vale:", i)
print()
print('El ciclo se ejecutó las veces requeridas, ojo que la variable iteradora siempre comienza en 0')
```

#### M8.2 Elabore un programa que cuente la cantidad de vocales que tiene una palabra ingresada por el usuario
1. El usuario debe ingresar
    - Palabra    


2. Reglas de negocio:
     - Se deben contar la cantidad de vocales que contiene la palabra

     
3. Mostrar la cantidad de vocales contenidas en una palabra
4. Recuerda válidar el ingreso de datos.

**Ejemplo de resultado**

Ingreso de datos:
    - Palabra: OUTLIERS

Mensaje:
- La palabra ingresada tiene 4 vocales

```python
#Desarrollo solución M8.2

p=input('Ingrese Palabra: ')
cont = 0


for vocal in p:
    if vocal == 'a' or vocal == 'e' or vocal == 'i' or vocal == 'o' or vocal == 'u':
        cont = cont+1 
           
          
print('La palabra', p, 'tiene', cont, 'vocales ')  

#10 min
    
```
